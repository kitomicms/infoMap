```python
import folium
from geopy.geocoders import Nominatim
import numpy as np
```

```python
geolocator = Nominatim(user_agent="infoMap")

def geolocate(country):
    try:
        # Geolocate the center of the country
        loc = geolocator.geocode(country)
        # And return latitude and longitude
        return (loc.latitude, loc.longitude)
    except:
        # Return missing value
        return np.nan

def map_generator(country,locations):
    x,y = geolocate(country)
    m = folium.Map(location=[x,y],zoom_start=6)
    for location in locations:
        location = "{},{}".format(location,country)
        #print(loc)
        x,y = geolocate(location)
        
        try:
            folium.Marker(
                [x,y], popup="<i>"+location+"</i>"
            ).add_to(m)
        except:
            pass
    return m
```
